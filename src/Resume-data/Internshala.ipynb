```python
import requests
import json
from bs4 import BeautifulSoup
```

```python
url = r"https://internshala.com/fresher-jobs/machine-learning-jobs/"
```

```python
response = requests.get(url)
```

```python
if response.status_code == 200:
    with open('Internshala_dataset.json', 'w') as json_file:
        json.dump(response.text, json_file, indent=4)
    print("Data saved to Internshala_dataset.json")
else:
    print(f"Error: {response.status_code} - {response.text}")
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
list_container = soup.find(id="list_container")
```

```python
jobs = list_container.find(id="internship_list_container")
```

```python
jobs = jobs.find('div', id="internship_list_container_1")
```

```python
div_tags = jobs.find_all('div', class_='individual_internship')
```

```python
count=1
for div in div_tags:
    if count==11:
        break
    field = div.find(class_='internship_meta')
    h3_tag = field.find('h3', class_='job-internship-name')
    a_tag = h3_tag.find('a', class_='job-title-href')
    job_tittle = a_tag.get_text(strip=True)
    href_link = a_tag.get('href')
    company_names = field.find('div', class_='company_and_premium')
    p_tag = company_names.find('p', class_='company-name')
    company_name = p_tag.get_text(strip=True)
    
    details_div = div.find('div', class_='individual_internship_details')
    location_tag = details_div.find('p', class_='locations')
    location = location_tag.get_text(strip=True) if location_tag else 'N/A'
    exp_icon = details_div.find('i', class_='ic-16-briefcase')
    experience = exp_icon.find_next('span').get_text(strip=True) if exp_icon else 'N/A'
    salary_icon = details_div.find('i', class_='ic-16-money')
    salary_desktop = salary_icon.find_next('span', class_='desktop').get_text(strip=True) if salary_icon else 'N/A'
    posted_div = details_div.find('div', class_='color-labels')
    posted_on = posted_div.find('span').get_text(strip=True) if posted_div else 'N/A'
    print(count, ".")
    count+=1
    print(f"Job Tittle: {job_tittle}")
    print(f"Company Name: {company_name}")
    print(f"Apply Link: https://internshala.com{href_link}")
    print(f"Job Location: {location}")
    print(f"Experience Required: {experience}")
    print(f"Salary: {salary_desktop}")
    print(f"Posted on: {posted_on}")
    print("\n\n\n\n")
    
```

```python
count=0
for job in jobs:
    count += 1
    print(count)
    title_tag = job.find('a', class_='job-title-href')
    title = title_tag.text.strip() if title_tag else 'N/A'
    print(f"Title       : {title}")

    company_tag = job.find('p', class_='company-name')
    company = company_tag.text.strip() if company_tag else 'N/A'
    print(f"Company     : {company}")

    location_tag = job.find('p', class_='row-1-item locations')
    location = location_tag.text.strip() if location_tag else 'N/A'
    print(f"Location    : {location}")

    exp_tag = job.find('div', class_='row-1-item')
    experience = exp_tag.find_next('span').text.strip() if exp_tag else 'N/A'
    print(f"Experience  : {experience}")

    salary_tag = job.find('span', class_='desktop')
    salary = salary_tag.text.strip() if salary_tag else 'N/A'
    print(f"Salary      : {salary}")

    posted_date_tag = job.find('div', class_='color-labels')
    posted_date = posted_date_tag.text.strip() if posted_date_tag else 'N/A'

    job_type_tag = job.find('div', class_='gray-labels').find('span')
    job_type = job_type_tag.text.strip() if job_type_tag else 'N/A'
    print(f"Job Type    : {job_type}")

    href = title_tag['href'] if title_tag else ''
    job_url = f'https://internshala.com{href}' if href else 'N/A'
    print(f"Job URL     : {job_url}")
    print()
print(count)
```

```python
count = 0
for job in jobs:
    count += 1
    title_tag = job.find('h3', class_='job-internship-name')
    title = title_tag.text.strip() if title_tag else 'N/A'

    company_tag = job.find('p', class_='company-name')
    company = company_tag.text.strip() if company_tag else 'N/A'

    logo_div = job.find('div', class_='internship_logo')
    logo_tag = logo_div.find('img') if logo_div else None
    logo_url = logo_tag['src'] if logo_tag else 'N/A'

    location_tag = job.find('p', class_='locations')
    location = location_tag.text.strip() if location_tag else 'N/A'

    stipend_div = job.find('div', class_='row-1-item')
    if stipend_div:
        stipend_tag = stipend_div.find_next_sibling('div').find_next_sibling('div')
        stipend = stipend_tag.text.strip() if stipend_tag else 'N/A'
    else:
        stipend = 'N/A'

    posted_date_tag = job.find('div', class_='color-labels')
    posted_date = posted_date_tag.text.strip() if posted_date_tag else 'N/A'
    
    href = title_tag['href'] if title_tag else ''
    job_url = f'https://internshala.com{href}' if href else 'N/A'

    print(f"{count}.")
    print(f"Title       : {title}")
    print(f"Company     : {company}")
    print(f"Logo URL    : {logo_url}")
    print(f"Location    : {location}")
    print(f"Stipend     : {stipend}")
    print(f"Posted Date : {posted_date}\n")
    print(f"Job URL     : {job_url}")
```
